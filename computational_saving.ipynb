```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
savings = [['PbPbPTB',188,329], ['AuAu',141,282],['PbPbCE',35,282]]
```

```python
savings_convrt = []
cpu_per_design = 1000
for name,tl, gp in savings:
    if name == 'AuAu':
        fac = 1000
        
    else:
        fac = 800
    savings_convrt.append([name, tl*fac,'TL'])
    savings_convrt.append([name, gp*fac,'GP'])


    
```

```python
savings_convrt
```

```python
saving_df = pd.DataFrame(data=savings_convrt,columns=['Target','CPU Hours','Emulation'])
```

```python
sns.set_context('poster')
fig,ax = plt.subplots(figsize=(10,10))
sns.barplot(data=saving_df,x='Target',y='CPU Hours',hue='Emulation',ax=ax, palette='bright')
plt.tight_layout()
fig.savefig('cpu_hours.png',dpi=100)
```
